# Define the predictor-corrector algorithm for this program (m=0)

```python
def PCS(beta,S,I,v,t):  #PC method for Susceptibile 
    Spre=S-(beta*S*I)*t #predictor
    Ipre=I+(beta*S*I-v*I)*t #predictor
    Scor=S-(beta*S*I+beta*Spre*Ipre)*t/2  #corrector
    return Scor

def PCI(beta,S,I,v,t):  #PC method for infected
    Spre=S-(beta*S*I)*t #predictor
    Ipre=I+(beta*S*I-v*I)*t  #predictor
    Icor=I+(beta*S*I-v*I+beta*Spre*Ipre-v*Ipre)*t/2  #corrector
    return Icor

def PCR(S,I,R,v,t): #PC method for recovered
    Rpre=R+(v*I)*t  #predictor
    Ipre=I+(beta*S*I-v*I)*t  #predictor
    Rcor=R+(v*I+v*Ipre)*t/2  #corrector
    return Rcor
```

#  Define the PC algorithm for m=0.001

```python
def PCSm(beta,S,I,v,m,t):  #PC method for Susceptibile 
    Spre=S-(beta*S*I-m+m*S)*t #predictor
    Ipre=I+(beta*S*I-v*I-m*I)*t #predictor
    Scor=S-(beta*S*I+beta*Spre*Ipre)*t/2+(2*m-m*S-m*Spre)*t/2  #corrector
    return Scor

def PCIm(beta,S,I,v,m,t):  #PC method for infected
    Spre=S-(beta*S*I+m-m*S)*t #predictor
    Ipre=I+(beta*S*I-v*I-m*I)*t  #predictor
    Icor=I+(beta*S*I-v*I+beta*Spre*Ipre-v*Ipre-m*I-m*Ipre)*t/2  #corrector
    return Icor

def PCRm(S,I,R,v,m,t): #PC method for recovered
    Rpre=R+(v*I-m*R)*t  #predictor
    Ipre=I+(beta*S*I-v*I-m*I)*t  #predictor
    Rcor=R+(v*I+v*Ipre-m*R-m*Rpre)*t/2  #corrector
    return Rcor
```
